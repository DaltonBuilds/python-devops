## List Comprehension

```python
servers = ["nginx-01", "nginx-02", "apache-01"]

servers_uppercase = [server.upper() for server in servers]

print(servers_uppercase)
```

```python
main_app_servers = ["frontend", "backend", "database", "monitoring"]

generate_ips = {server: f"192.168.10.{i}" for i, server in enumerate(main_app_servers, 10)}

print(generate_ips)

```

```python
roles = ['primary-db', 'replica-db', 'cache']
hosts = ['db01.prod', 'db02.prod']
 
server_map = {role: host for role, host in zip(roles, hosts)}

print(server_map)
```

```python
filenames = ["PROD.YML", "settings.json", "app.yaml", "DEV.YML", "conftest.py"]

lowercase_filenames = [f.lower() for f in filenames if f.lower().endswith(('.yml', 'yaml'))]

print(lowercase_filenames)
```
