## Guard Clauses
*Handle edge cases at the top of functions to avoid deep nesting `if` conditions.*

```python
def process_data_guarded(data):
    if not data:
        print("No data provided")

    elif not isinstance(data, list):
        print(f"Data must be a list. Provided: {type(data)}")
        return 

    else:
        print(f"Processing {len(data)} items...")
        print("Data processed successfully!")

process_data_guarded([])
process_data_guarded(123)
process_data_guarded(None)
process_data_guarded([1, 2, 3]) # only successful call
process_data_guarded("Hello, world!")

```
