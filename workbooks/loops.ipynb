```python
connection_attempts = 0
max_attempts = 7
connected = False

while not connected and connection_attempts < max_attempts:
    print(f"Attempting to connect... ({connection_attempts + 1}/{max_attempts})")

    if connection_attempts == 6: # Simulate a connection attempt
        connected = True

    connection_attempts += 1

if not connected:
    print(f"Failed to connect after {max_attempts} attempts.")
else:
    print("Connection successful!")
```
